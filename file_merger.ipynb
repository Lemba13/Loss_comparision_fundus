```python
import os
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

#### This notebook merges all the available augmentated images into a single numpy file for easier iteration during training. 

```python
path = '/content/drive/MyDrive/DRIVE project/dataset/augmentation'
```

```python
X = list()
Y = list()
for i in os.listdir(path):
  for j in os.listdir(path+'/'+i):
    if j == 'images':
      for k in sorted(os.listdir(path+'/'+i+'/'+j)):
        impathX = path+'/'+i+'/'+j+'/'+ k
       # X.append(cv2.resize(cv2.cvtColor(cv2.imread(impathX), cv2.COLOR_BGR2GRAY), (512,512), interpolation = cv2.INTER_NEAREST))
        X.append(cv2.cvtColor(cv2.imread(impathX), cv2.COLOR_BGR2GRAY))
    elif j == 'labels':
      for k in sorted(os.listdir(path+'/'+i+'/'+j)):
        impathY = path+'/'+i+'/'+j+'/'+ k
        #Y.append(cv2.resize(cv2.cvtColor(cv2.imread(impathY), cv2.COLOR_BGR2GRAY ), (512,512), interpolation = cv2.INTER_NEAREST))
        Y.append(cv2.cvtColor(cv2.imread(impathY), cv2.COLOR_BGR2GRAY ))
```

```python
X = np.array(X)
Y = np.array(Y)
print(Y.shape)
```

```python
unique,counts = np.unique(Y.flatten(), return_counts = True)
import matplotlib.pyplot as plt
plt.figure(figsize = (10,8))
plt.scatter(unique,counts, color = 'red')
```

```python
for i in range(len(unique)):
  if i%20 == 0:
    print(f'Unique values:{unique[i]} and its frequency:{counts[i]}')
  if i == 255:
    print(f'Unique values:{unique[i]} and its frequency:{counts[i]}')
```

```python
PATH = '/content/drive/MyDrive/DRIVE project/dataset/test'
test_images = list();
test_labels = list();
for i in os.listdir(PATH):
  if i == 'images':
    for j in sorted(os.listdir(PATH+'/'+str(i))):
      print(j)
      impathX = PATH+'/'+str(i)+'/'+str(j)
      test_images.append(cv2.resize(cv2.cvtColor(cv2.imread(impathX), cv2.COLOR_BGR2GRAY), (512,512), interpolation = cv2.INTER_NEAREST))
  if i == 'labels':
    for j in sorted(os.listdir(PATH+'/'+str(i))):
      print(j)
      impathY = PATH+'/'+str(i)+'/'+str(j)
      test_labels.append(cv2.resize(cv2.cvtColor(cv2.imread(impathY), cv2.COLOR_BGR2GRAY), (512,512), interpolation = cv2.INTER_NEAREST))
```

```python
test_images = np.array(test_images)
test_labels = np.array(test_labels)
test_images = np.expand_dims(test_images, axis = 3)
test_labels = np.expand_dims(test_labels, axis = 3)
```
